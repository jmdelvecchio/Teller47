```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.colors as colors
```

```python
df = pd.read_csv('T47_GCPs_extract.csv',na_values='-9999')

topos = df.columns[-4:]
diffs = ['dElev_1819', 'dDist_1819']
```

```python
df
```

```python
slope_normal = ((df[diffs[0]])**2 + df[diffs[1]]**2)**0.5
```

```python
#fig, axes = plt.subplots(2,4,figsize=(7,3.5), sharey=True)
fig, axes = plt.subplots(2,4,figsize=(7,3.5))
#Though I want to have different values for the maximum of elevation vs position...

for i, ax in enumerate(axes[0]):
#     ax = sns.regplot(x=df[topos[i]],y=df[diffs[0]])
    ax.scatter(x=df[topos[i]],y=(df[diffs[0]]))

for i, ax in enumerate(axes[1]):
#     ax = sns.regplot(x=df[topos[i]],y=df[diffs[0]])
    ax.scatter(x=df[topos[i]],y=(df[diffs[1]]))  
    ax.set_xlabel(topos[i], fontsize=12)

axes[0,0].set_ylabel('d elevation (m)', fontsize=12)
#axes[0,0].set_ylim(-0.05, 0.2)
axes[1,0].set_ylabel('d position (m)', fontsize=12)
#axes[1,0].set_ylim(-0.05, 0.3)
axes[0,0].set_xscale('log')
axes[0,1].tick_params(labelleft=False) #Finish this in a loop!
fig.suptitle('Average annual motion of GCPs', fontsize=12)
#fig.tight_layout()
plt.savefig("GCPs_vs_topos.pdf")
```

```python
fig, axes = plt.subplots(1,4,figsize=(7,2), sharey=True)
#fig, axes = plt.subplots(2,4,figsize=(7,3.5))
#Though I want to have different values for the maximum of elevation vs position...

for i, ax in enumerate(axes):
#     ax = sns.regplot(x=df[topos[i]],y=df[diffs[0]])
    ax.scatter(x=df[topos[i]],y=(slope_normal))
```

```python
sns.regplot(data=df,x='radiation',y=slope_normal)

```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(7,3), sharey=True)
im = ax1.scatter(x=df['slope'],y=slope_normal,c=df['flowac3m_1'],cmap='PuBu',norm=colors.LogNorm(), edgecolor='black', s=50)
cb = fig.colorbar(im, ax=ax1)
ax1.set_title('Colored by drainage area (m2)')
im = ax2.scatter(x=df['slope'],y=slope_normal,c=df['radiation'],cmap='gnuplot2', edgecolor='black', s=50)
cb = fig.colorbar(im, ax=ax2)
ax2.set_title('Colored by insolation (W/m2)');
fig.suptitle('Slope versus total movement, GCPs', fontsize=16, y=1.08)
ax1.set_ylabel('Total movement, m \n (vertical and horizontal)')
#fig.tight_layout()
fig.savefig('GCPs_color_scatters.svg', dpi=200)
```

```python
d17_19 = df2.columns[4:6]
d17_18 = df2.columns[7:9]
d18_19 = df2.columns[10:12]
```

```python
fig, ax = plt.subplots(2,3,figsize=(7,5))
ax[0,2].scatter(x=df2[d17_19[0]],y=df2[d17_19[1]])
ax[0,1].scatter(x=df2[d17_18[0]],y=df2[d17_18[1]])
ax[0,0].scatter(x=df2[d18_19[0]],y=df2[d18_19[1]])
ax[1,0].scatter(x=df2[d17_18[0]],y=df2[d18_19[0]])
ax[1,1].scatter(x=df2[d17_18[1]],y=df2[d18_19[1]])
ax[1,2].scatter(x=(df2[d18_19[1]]/df2[d17_19[1]]),y=df2[d17_19[1]])

ax[0,2].set_xlabel('Elevation change 17-19')
ax[0,1].set_xlabel('Elevation change 17-18')
ax[0,0].set_xlabel('Elevation change 18-19')
ax[1,0].set_xlabel('Elevation change 17-18')
ax[1,1].set_xlabel('Position change 17-18')
ax[1,2].set_ylabel('Position change total')

ax[0,2].set_ylabel('Position change 17-19')
ax[0,1].set_ylabel('Position change 17-18')
ax[0,0].set_ylabel('Position change 18-19')
ax[1,0].set_ylabel('Elevation change 18-19')
ax[1,1].set_ylabel('Position change 18-19')
ax[1,2].set_xlabel('% Position change\n18-19 of total change')

fig.suptitle('Interannual comparison of lobe targets', fontsize=12)
fig.tight_layout()
fig.savefig('lobe_crossplots.pdf', dpi=200)
```
